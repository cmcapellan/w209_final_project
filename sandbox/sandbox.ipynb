```python
import json

with open('astronaut.json') as f:
    data = json.load(f)

gender_group = {}

for key in data:
    if data[key]["Group"] not in gender_group.keys():
        gender_group[data[key]["Group"]] = {"M":0, "W":0, "Mil":0, "Civ":0}
    if data[key]["Gender"] == "Male":
        gender_group[data[key]["Group"]]["M"] += 1
    elif data[key]["Gender"] == "Female":
        gender_group[data[key]["Group"]]["W"] += 1
    if data[key]["Military Branch"] == "":
        gender_group[data[key]["Group"]]["Civ"] += 1
    elif data[key]["Military Branch"] != "":
        gender_group[data[key]["Group"]]["Mil"] += 1

#print(gender_group)
ord_gen_grp = {}
final_gen_grp = []

for i in range(1,23):
    ord_gen_grp[i] = gender_group[i]
    ord_gen_grp[i]["pctM"] = ord_gen_grp[i]["M"]/(ord_gen_grp[i]["M"] + ord_gen_grp[i]["W"])
    ord_gen_grp[i]["pctW"] = ord_gen_grp[i]["W"]/(ord_gen_grp[i]["M"] + ord_gen_grp[i]["W"])
    ord_gen_grp[i]["pctMil"] = ord_gen_grp[i]["Mil"]/(ord_gen_grp[i]["Mil"] + ord_gen_grp[i]["Civ"])
    ord_gen_grp[i]["pctCiv"] = ord_gen_grp[i]["Civ"]/(ord_gen_grp[i]["Mil"] + ord_gen_grp[i]["Civ"])
    final_gen_grp.append({"Group": i})
    final_gen_grp[i-1]["pctM"] = ord_gen_grp[i]["M"]/(ord_gen_grp[i]["M"] + ord_gen_grp[i]["W"])
    final_gen_grp[i-1]["pctW"] = ord_gen_grp[i]["W"]/(ord_gen_grp[i]["M"] + ord_gen_grp[i]["W"])
    final_gen_grp[i-1]["W"] = ord_gen_grp[i]["W"]
    final_gen_grp[i-1]["M"] = ord_gen_grp[i]["M"]
    final_gen_grp[i-1]["pctMil"] = ord_gen_grp[i]["Mil"]/(ord_gen_grp[i]["Mil"] + ord_gen_grp[i]["Civ"])
    final_gen_grp[i-1]["pctCiv"] = ord_gen_grp[i]["Civ"]/(ord_gen_grp[i]["Mil"] + ord_gen_grp[i]["Civ"])
    final_gen_grp[i-1]["pctMil"] = ord_gen_grp[i]["pctMil"]
    final_gen_grp[i-1]["pctCiv"] = ord_gen_grp[i]["pctCiv"]
    
    
#print(ord_gen_grp)
print(final_gen_grp)


with open('gen_grp.json', 'w') as json_file:
    json.dump(ord_gen_grp, json_file)

with open('gen_grp.txt', 'w') as json_file:
    json.dump(final_gen_grp, json_file)
```

```python
import csv

missions = []
with open('Space_Corrected.csv', mode='r', encoding='utf-8') as infile:
    reader = csv.DictReader(infile)
    for i, row in enumerate(reader):
        missions.append({"Row": i})
        missions[i]['Datum'] = row['Datum']
        missions[i]['Detail'] = row['Detail']
        merc_list = ["MR-3","MR-4","MA-6","MA-7","MA-8","MA-9"]
        for item in merc_list:
            if item in row['Detail']:
                missions[i]['Program'] = "Mercury"
        gem_list = ["Gemini 3","Gemini IV","Gemini V","Gemini VI-A","Gemini VII","Gemini VIII",\
                    "Gemini IX-A","Gemini X","Gemini XI","Gemini XII"]
        for item in gem_list:
            if item in row['Detail']:
                missions[i]['Program'] = "Gemini"
        apo_list = ["Apollo 7","Apollo 9","Apollo 8","Apollo 10","Apollo 11","Apollo 12","Apollo 13",\
                    "Apollo 14","Apollo 15","Apollo 16","Apollo 17"]
        for item in apo_list:
            if item in row['Detail']:
                missions[i]['Program'] = "Apollo"
        sky_list = ["Skylab 2","Skylab 3","Skylab 4"]
        for item in sky_list:
            if item in row['Detail']:
                missions[i]['Program'] = "Skylab"
        if "space shuttle columbia" in row['Detail'].lower():
            missions[i]['Program'] = "Space Shuttle Columbia"
        if "space shuttle challenger" in row['Detail'].lower():
            missions[i]['Program'] = "Space Shuttle Challenger"
        if "space shuttle discovery" in row['Detail'].lower():
            missions[i]['Program'] = "Space Shuttle Discovery"
        if "space shuttle atlantis" in row['Detail'].lower():
            missions[i]['Program'] = "Space Shuttle Atlantis"
        if "space shuttle endeavour" in row['Detail'].lower():
            missions[i]['Program'] = "Space Shuttle Endeavour"
        iss_list = ["Soyuz TM-31","STS-102","STS-105","STS-108","STS-111","STS-113","Soyuz TMA-2","Soyuz TMA-3",\
                    "Soyuz TMA-4","Soyuz TMA-5","Soyuz TMA-6","Soyuz TMA-7","Soyuz TMA-8","STS-121","Soyuz TMA-9",\
                    "STS-116","Soyuz TMA-10","STS-117","Soyuz TMA-11","STS-120","STS-122","STS-123","Soyuz TMA-12",\
                    "STS-124","Soyuz TMA-13","STS-126","STS-119","Soyuz TMA-14","STS-127","Soyuz TMA-15","STS-128",\
                    "Soyuz TMA-16","Soyuz TMA-17","Soyuz TMA-18","Soyuz TMA-19","Soyuz TMA-01M","Soyuz TMA-20",\
                    "Soyuz TMA-21","Soyuz TMA-02M","Soyuz TMA-22","Soyuz TMA-03M","Soyuz TMA-04M","Soyuz TMA-05M",\
                    "Soyuz TMA-06M","Soyuz TMA-07M","Soyuz TMA-08M","Soyuz TMA-09M","Soyuz TMA-10M","Soyuz TMA-11M",\
                    "Soyuz TMA-12M","Soyuz TMA-13M","Soyuz TMA-14M","Soyuz TMA-15M","Soyuz TMA-16M","Soyuz TMA-17M",\
                    "Soyuz TMA-18M","Soyuz TMA-19M","Soyuz TMA-20M","Soyuz MS-01","Soyuz MS-02","Soyuz MS-03",\
                    "Soyuz MS-04","Soyuz MS-05","Soyuz MS-06","Soyuz MS-07","Soyuz MS-08","Soyuz MS-09","Soyuz MS-11",\
                    "Soyuz MS-12","Soyuz MS-13","Soyuz MS-15","Soyuz MS-16","SpaceX DM-2","Soyuz MS-17"]
        for item in iss_list:
            if item in row['Detail']:
                missions[i]['Program'] = "ISS"

i = len(missions) - 1 
for row in reversed(missions):
    if 'Program' not in row.keys():
        del missions[i]
    i -= 1

with open("missions.txt", "w") as text_file:
    text_file.write(str(missions))
```
