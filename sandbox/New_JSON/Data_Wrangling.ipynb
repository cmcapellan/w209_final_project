```python
import json

with open('astronaut.json') as f:
    data = json.load(f)
    
one_row_per_ast = []
undergrad = []
grad = []
alma_mater = []
missions = []
```

```python
for i, ele in enumerate(data):
    data[i]['index'] = i
    missions.append({'index':i,'Name':data[i]['Name'],'Missions':data[i].pop('Missions')})
    alma_mater.append({'index':i,'Name':data[i]['Name'],'Alma Mater':data[i].pop('Alma Mater')})
    undergrad.append({'index':i,'Name':data[i]['Name'],'Undergraduate Major':data[i].pop('Undergraduate Major')[0].split('; ')})
    grad.append({'index':i,'Name':data[i]['Name'],'Graduate Major':data[i].pop('Graduate Major')[0].split('; ')})
    if data[i]['Military Branch'] == '':
        data[i]['Military Branch'] = 'Civilian'
    
with open('astronaut_orpa.json', 'w') as json_file:
    json.dump(data, json_file)
```

```python
with open('missions.json', 'w') as json_file:
    json.dump(missions, json_file)
```

```python
with open('alma_mater.json', 'w') as json_file:
    json.dump(alma_mater, json_file)
```

```python
for i in undergrad:
    i['Maj Cat']=[]
    for m in i['Undergraduate Major']:
        if m == '':
            i['Maj Cat'].append("No Graduate Degree")
        elif "engineer" in m.lower():
            if "elect" in m.lower():
                i['Maj Cat'].append("Electrical Engineering")
            elif "astro" in m.lower():
                i['Maj Cat'].append("Astronautical/Aerospace Engineering")
            elif "aero" in m.lower():
                i['Maj Cat'].append("Astronautical/Aerospace Engineering")
            elif "mech" in m.lower():
                i['Maj Cat'].append("Mechanical Engineering")
            else:
                i['Maj Cat'].append("Other Engineering")
        elif "physics" in m.lower():
            i['Maj Cat'].append("Physics")
        elif "medicine" in m.lower():
            i['Maj Cat'].append("Medicine")
        elif "health" in m.lower():
            i['Maj Cat'].append("Medicine")
        elif "business" in m.lower():
            i['Maj Cat'].append("Business")
        else:
            i['Maj Cat'].append(m)
undergrad
```

```python
with open('undergrad.json', 'w') as json_file:
    json.dump(undergrad, json_file)
```
